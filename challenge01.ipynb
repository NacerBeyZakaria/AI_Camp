```python
# Cash Dicounts
value = int(input("Enter The Cash value:"))
if value >= 1000:
    discount = "20%"
    value = value - value*20/100
elif  value <1000 and value >= 500 :
    discount = "10%"
    value = value - value*10/100
else :
    discount ="0%"
print(f"The new value is {value} and discount is {discount}")
```

```python
# Largest Word
def largest_word(sentence):
    words = sentence.split()
    large_word = words[0]

    for word in words:
        if len(word) > len(large_word):
            large_word = word


    return large_word

sentence = input("Enter a sentence:")
LW = largest_word(sentence)
print(f"Largest Word Is :{LW}")
```

```python
# Average of the numbers in a given list
def average_number(list):
    i= 0
    summ = 0
    for num in list:
          n = int(num)
          i += 1
          summ = summ + n
    if i != 0:
         return summ/i

number = int(input("Enter the number of the numbers that you want to add :"))
i = 0
list1 = []
while i < number :
     temp_num = input(f"enter the {i+1} number :")
     list1.append(temp_num)
     i = i+ 1
z = average_number(list1)
print(f"The average is : {z}")
```
