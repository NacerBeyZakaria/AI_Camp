```python
# Challenge 1 :
import numpy as np
import time

# 1. Create two large random vectors
v1 = np.random.rand(1000000)
v2 = np.random.rand(1000000)

# 2. The function
def dot_product(v1,v2):
    dot_pro = 0
    if len(v1) != len(v2) :
        print("The vectors must be with the same size !!")
        return
    else :
        i = 0
        while i < len(v1):
            dot_pro += v1[i]*v2[i]
            i += 1
        return dot_pro


start_time_loop = time.time()

z = dot_product(v1,v2)
print(f"The dot product of {v1} and {v2} is : {z}")
end_time_loop = time.time()
execution_time_loop = end_time_loop - start_time_loop
print(f"The function takes :{execution_time_loop}")

start_time_loop = time.time()
result = np.dot(v1, v2)
print(f"The dot product of {v1} and {v2} is : {result}")
execution_time_loop = end_time_loop - start_time_loop
print(f"The build in numpy function takes :{execution_time_loop}")

# The diffrenece in execution time is huge !! , in the build in dot function is faster, i think due to its complexity (maybe o(n) or less)
```

```python
# Challenge 2 :
import sympy as sp
neta = 0.1

def gradient_descent(f, f_derivative, x0=10, tolerance=1e-6, max_iterations=100):
    iteration = 0
    while iteration < max_iterations:
        grad = f_derivative(x0)
        if abs(grad) < tolerance:
            return x0, f(x0)
        x0 = x0 - neta * grad
        if iteration % 10 == 0:
          print(f"after {iteration} iterations, the x value is :{x0} ,and the f(x) value is :{f(x0)}")
        iteration += 1
    return x0, f(x0)

def get_function_and_derivative_from_user():
    x = sp.symbols('x')
    user_input = input("Enter a function of x :")

    func = sp.sympify(user_input)
    func_derivative = sp.diff(func, x)

    f = sp.lambdify(x ,func ,'numpy')
    f_derivative = sp.lambdify(x ,func_derivative, 'numpy')

    return f, f_derivative

f,f_derivative = get_function_and_derivative_from_user()

minimum_x, minimum_value = gradient_descent(f, f_derivative)
print("Minimum x:", minimum_x)
print("Minimum value:", minimum_value)
```

```python
# Challenge 3:
import pandas as pd
from google.colab import files

uploaded = files.upload()
df = pd.read_csv("cereal.csv")

dataset = pd.DataFrame(df)

# 2. Calculate mean and median for calories, protein, sugars
print("Mean & Median Values:")
for col in ['calories', 'protein', 'sugars']:
    mean_val = df[col].mean()
    median_val = df[col].median()
    print(f"  {col.capitalize()} - Mean: {mean_val:.2f}, Median: {median_val}")

# 3. Identify outliers in 'calories' column
Q1 = df['calories'].quantile(0.25)
Q3 = df['calories'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['calories'] < lower_bound) | (df['calories'] > upper_bound)]['calories']

print("\nCalories Outlier Detection:")
print(f"  Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
print(f"  Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
print(f"  Outlier calories: {outliers.tolist()}")

# 4. Compare mean and median for sugars
sugars_mean = df['sugars'].mean()
sugars_median = df['sugars'].median()
interpret = (
    "right-skewed (many low, some very high sugars)" if sugars_mean > sugars_median else
    "left-skewed (many high, few low sugar cereals)" if sugars_mean < sugars_median else
    "fairly symmetric"
)
print(f"\nSugars Distribution: {interpret}")
```
